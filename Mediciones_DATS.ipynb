# Gráficas para las mediciones realizadas con el DATS

## Gráficas comparativas de $|Z_e|$ y su fase

```python
#Función para graficar las impedancia eléctrica y su fase

def graficar_impedancia_fase_lado_a_lado(archivos, etiquetas, categorias_color,
                                         titulo,
                                         lim_x=None,
                                         lim_y_imp=None,
                                         lim_y_phase=None,
                                         nombre_archivo=None):

    import os
    import pandas as pd
    import matplotlib.pyplot as plt
    import numpy as np

    fig, (ax_imp, ax_phase) = plt.subplots(1, 2, figsize=(16, 6), sharex=True)

    for archivo, etiqueta in zip(archivos, etiquetas):
        df = pd.read_csv(os.path.join('dats', archivo))
        color = categorias_color.get(etiqueta, '#000000')

        # Impedancia
        ax_imp.plot(df['Freq'], df['Impedance'], label=etiqueta, color=color)

        # Fase
        ax_phase.plot(df['Freq'], df['Phase'], linestyle='--', label=etiqueta, color=color)

    # Configurar eje de impedancia
    ax_imp.set_xscale('log')
    ax_imp.set_xlim(lim_x)
    ax_imp.set_ylim(lim_y_imp)
    ax_imp.set_xlabel('Frecuencia [Hz]')
    ax_imp.set_ylabel('ZE [Ohms]')
    ax_imp.set_title('Impedancia')
    ax_imp.grid(True, which='both', linestyle=':', alpha=0.4)
    ax_imp.legend(loc='upper right')

    # Configurar eje de fase
    ax_phase.set_xscale('log')
    ax_phase.set_xlim(lim_x)
    ax_phase.set_ylim(lim_y_phase)
    ax_phase.set_xlabel('Frecuencia [Hz]')
    ax_phase.set_ylabel('Fase [Grados]')
    ax_phase.set_title('Fase')
    ax_phase.grid(True, which='both', linestyle=':', alpha=0.4)
    ax_phase.legend(loc='lower right')

    fig.suptitle(titulo, fontsize=16)
    plt.grid(True)
    plt.tight_layout(rect=[0, 0, 1, 0.95])

# Guardar la figura si se proporciona nombre
    if nombre_archivo:
        os.makedirs('figuras', exist_ok=True) #Las figuras se guardaran en una carpeta llama "figuras", sino la encuentra la creará.
        plt.savefig(os.path.join('figuras', f'{nombre_archivo}.png'), dpi=300)

    plt.show()
```

#### Los siguientes gráficos están comparados como:
- Mediciones sin carbón activado: Contiene las mediciones realizadas con...
  - Parlante Solo
  - Caja Acústica
  - Caja + Peltier
- Mediciones con 1 cm de carbón activado: Contiene las mediciones realizadas con...
  - Carbón 1cm Sin Tratar
  - Carbón 1cm Sin Tratar + Peltier
  - Carbón 1cm Tratado
  - Carbón 1cm Tratado + Peltier
- Mediciones con 2 cm de carbón activado: Contiene las mediciones realizadas con...
  - Carbón 2cm Sin Tratar
  - Carbón 2cm Sin Tratar + Peltier
  - Carbón 2cm Tratado + Peltier
- Mediciones con 3 cm de carbón activado: Contiene las mediciones realizadas con...
  - Carbón 3cm Sin Tratar
  - Carbón 3cm Sin Tratar + Peltier
  - Carbón 3cm Tratado + Peltier
